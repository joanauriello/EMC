```python
import rpy2.robjects as ro
from rpy2.robjects import r
```

```python
# Load the RData file
ro.r['load']('mundodes.RData')
```

```python
print(r['ls']())
```

```python
# Extract the R object 'mundodes'
mundodes = ro.r['mundodes']

# Write the R object to a CSV file with row names
ro.r('write.csv(mundodes, "mundodes.csv", row.names=TRUE)')
```
